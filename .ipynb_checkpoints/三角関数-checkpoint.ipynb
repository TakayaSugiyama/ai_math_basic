# 三角関数

```python
import numpy as np
import matplotlib.pyplot as plt 

def my_sin(x):
    return np.sin(x)   #sin(θ)

def my_cos(x):
    return np.cos(x)  #cos(θ)

x = np.linspace(-np.pi, np.pi)
y_sin = my_sin(x)
y_cos = my_cos(x)

plt.plot(x,y_sin,label = "sin(x)")
plt.plot(x,y_cos,label = "cos(x)")

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 

def my_tan(x):
    return np.tan(x)

x = np.linspace(-1.3,1.3)
y = my_tan(x)

plt.plot(x,y,label = "tan(x)")
plt.legend()
plt.show()
```
