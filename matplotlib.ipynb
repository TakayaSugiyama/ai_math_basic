## linspace関数

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(-5,5)
y = 2 * x

plt.plot(x,y)
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

x =np.linspace(-5,5)
y_1 = 2 * x
y_2 = 3 * x

plt.xlabel("x")
plt.ylabel("y")
           
plt.title("My Graph")
plt.plot(x,y_1,label= "y1",linestyle = "dashed")
plt.plot(x,y_2,label ="y2")
           
plt.legend()
plt.show()
```

## 散布図の表示

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.array([1.2, 2.4, 0.0 ,1.4, 1.5])
y = np.array([2.4,1.4, 1.0, 0.1, 1.7])

plt.scatter(x,y)
plt.show()
```
