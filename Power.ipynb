$$ y = a^x $$

```python
import numpy as np
import matplotlib.pyplot as plt 

def my_func(x):
    return 3 ** x

x = np.linspace(0,2)
y = my_func(x)
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 

def my_func(x):
    return np.sqrt(x)  #xの平方根

x = np.linspace(0,9)
y = my_func(x)
plt.plot(x,y)
plt.show()
```
