# 確率の概念

## 確率への収束

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = []
y = []
total = 0
num_5 = 0 
n = 5000 

for i in range(n):
    if np.random.randint(6) + 1 == 5:
        num_5 += 1
    total += 1
    x.append(i)
    y.append(num_5/total) 

plt.plot(x,y)
plt.plot(x, [1/6]*n, linestyle = "dashed")
plt.show()
```

## コインの表裏の確率が$\frac{1}{2}$になる確率

```python
import numpy as numpy
import matplotlib.pyplot as plt  

x = []
y = []
total = 0  #試行数
num_front = 0 #表が上になった回数
n = 5000 #コインを投げる回数

for i in range(n):
    if np.random.randint(2) == 0:
        num_front += 1
    total += 1
    x.append(i)
    y.append(num_front / total)

plt.plot(x,y)
plt.plot(x,[1/2]*, linestyle = "dashed")
plt.show()
```
