$$ y = 3x^2 + 2x + 1 $$

```python
import numpy as np
import matplotlib.pyplot as plt 

def my_func(x):
    return 3*x**2 + 2*x + 1

x = np.linspace(-5,5)
y = my_func(x)
plt.plot(x,y)
plt.show()
```

$$ y = 4x^3 - 2x^2 + x + 1 $$

```python
import numpy as np
import matplotlib.pyplot as plt 

def my_func(x):
    return 4*x**3 - 2*x**2 + x + 1

x = np.linspace(-5,5)
y = my_func(x)

plt.plot(x,y)
plt.show()
```
